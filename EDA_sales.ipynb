# **Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Import the data**

```python
df_stores=pd.read_csv(r'E:\FinalProject\data\stores_data_set.csv')
df_stores.head()
```

```python
df_features=pd.read_csv(r'E:\FinalProject\data\Features_data_set.csv')
df_features
```

```python
df_sales=pd.read_csv(r'E:\FinalProject\data\sales_data_set.csv')
df_sales
```

# **Merge the data in a unique DataFrame**


```python
df_stores_features=df_stores.merge(right=df_features,on='Store')

df_merge=df_stores_features.merge(right=df_sales,on=['Store', 'Date', 'IsHoliday'])
df_merge.sample(10)
```

# **Checking Data and feature**

```python
df_merge.info()
```

```python
df_merge.describe()
```

```python
df_merge['Date'] = pd.to_datetime(df_merge['Date'],format='%d/%m/%Y')
```

```python
df_temp = df_merge.copy(deep=True)
df_temp['MarkDown'] = df_temp['MarkDown1'] + df_temp['MarkDown2'] +df_temp['MarkDown3'] +df_temp['MarkDown4'] + df_temp['MarkDown5']
df_temp.drop(['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5'], inplace = True, axis = 1)
```
